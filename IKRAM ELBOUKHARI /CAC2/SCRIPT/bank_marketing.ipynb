## Visualize Numerical Features

### Subtask:
Create histograms for key numerical features like 'age' and 'balance' to show their distributions.


**Reasoning**:
I need to import the necessary libraries, create a figure with two subplots, generate histograms for 'age' and 'balance' as specified in the instructions, and then display the plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for 'age'
sns.histplot(X['age'], kde=True, ax=axes[0])
axes[0].set_xlabel('Age')
axes[0].set_title('Distribution of Age')

# Histogram for 'balance'
sns.histplot(X['balance'], kde=True, ax=axes[1])
axes[1].set_xlabel('Balance (Euros)')
axes[1].set_title('Distribution of Balance')

plt.tight_layout()
plt.show()
```

## Visualize Categorical Features

### Subtask:
Generate bar plots for important categorical features such as 'job' and 'marital' to illustrate the frequency of each category. Each plot will include appropriate labels, titles, and a legend if applicable.


**Reasoning**:
To visualize the distribution of categorical features, I will create bar plots for 'job' and 'marital' columns using `seaborn.countplot()`, displaying them side-by-side in subplots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar plot for 'job'
sns.countplot(data=X, y='job', ax=axes[0], palette='viridis')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Job Type')
axes[0].set_title('Distribution of Job Types')

# Bar plot for 'marital'
sns.countplot(data=X, y='marital', ax=axes[1], palette='plasma')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Marital Status')
axes[1].set_title('Distribution of Marital Status')

plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code generated `FutureWarning` messages because `palette` was used without assigning `hue` in `sns.countplot`. To fix this and follow the recommended practice for newer seaborn versions, I will explicitly assign the 'job' and 'marital' columns to the `hue` parameter and set `legend=False` to maintain the desired aesthetic without the warning.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar plot for 'job'
sns.countplot(data=X, y='job', ax=axes[0], hue='job', legend=False, palette='viridis')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Job Type')
axes[0].set_title('Distribution of Job Types')

# Bar plot for 'marital'
sns.countplot(data=X, y='marital', ax=axes[1], hue='marital', legend=False, palette='plasma')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Marital Status')
axes[1].set_title('Distribution of Marital Status')

plt.tight_layout()
plt.show()
```

## Visualize Target Variable

### Subtask:
Create a bar plot for the target variable 'y' to show the distribution of 'yes' and 'no' responses, providing insight into the class imbalance. The plot will include a title and labels.


**Reasoning**:
To visualize the distribution of the target variable 'y' and assess class imbalance, I will create a bar plot using `seaborn.countplot`, ensuring to set appropriate labels and a title as per the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(data=y, x='y', palette='coolwarm')
plt.xlabel('Subscription Status')
plt.ylabel('Count')
plt.title('Distribution of Target Variable (y)')
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because `palette` was used without assigning `hue` in `sns.countplot`. To address this, I will explicitly set the `hue` parameter to 'y' and `legend=False` to follow the recommended practice for newer seaborn versions, similar to the fix applied for categorical features, while still maintaining the desired visualization.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(data=y, x='y', hue='y', legend=False, palette='coolwarm')
plt.xlabel('Subscription Status')
plt.ylabel('Count')
plt.title('Distribution of Target Variable (y)')
plt.show()
```
